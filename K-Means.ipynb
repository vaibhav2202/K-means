```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples = 500,n_features = 2)
```

```python
plt.grid(True)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
k = 3

color = ['green', 'red', 'yellow', 'blue', 'orange']

clusters = {}
for i in range(k):
    centre = 10*(2*np.random.random((x.shape[1],))-1)
    points = []

    cluster = {
        'center': centre,
        'points':points,
        'color':color[i]
    }
    
    clusters[i] = cluster
```

```python
clusters
```

```python
def distance(v1, v2):
    return np.sqrt(np.sum((v1-v2)**2))
```

```python
distance(5.2133,12.3466)
```

```python
def getClusters(clusters):
    for ix in range(x.shape[0]):
        dis = []
        current_cl = x[ix]
        
        for kx in range(k):
            d= distance(current_cl, clusters[kx]['center'])
            dis.append(d)
        val = np.argmin(dis)
        
        clusters[val]['points'].append(current_cl)
```

```python
def updateClusters(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        
        if pts.shape[0] > 0:
            new_u = pts.mean(axis = 0)
            clusters[kx]['center'] = new_u
            clusters[kx]['points'] = []
```

```python
def plotClusters(clusters):
    for kx in range(k):
        pts = np.array(clusters[kx]['points'])
        print(pts.shape)
        
        try:
            plt.scatter(pts[:,0],pts[:,1], color = clusters[kx]['color'])
        except:
            pass
        uk = clusters[kx]['center']
        plt.scatter(uk[0], uk[1], color = 'black')
```

```python
getClusters(clusters)
plotClusters(clusters)
updateClusters(clusters)
```
